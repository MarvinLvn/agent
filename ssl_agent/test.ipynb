```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from matplotlib.patches import FancyArrowPatch
from matplotlib.legend_handler import HandlerLine2D

class DoubleArrowHandler(HandlerLine2D):
    def create_artists(self, legend, orig_handle, xdescent, ydescent, width, height, fontsize, trans):
        x = width - xdescent
        y = height / 2. - ydescent
        arrow = FancyArrowPatch((0, y), (x, y), 
                                arrowstyle='<->', 
                                color=orig_handle.get_color(),
                                mutation_scale=10,
                                linewidth=2)
        return [arrow]

# Define colors and labels
colors = ['red', 'blue', 'green', 'orange', 'purple', 'brown']
labels = ['Jaw height', 'Tongue dorsum', 'Lip protrusion', 
          'Tongue body', 'Tongue tip', 'Lip height']

# Create a figure and axis
fig, ax = plt.subplots(figsize=(6, 3))  # Adjust size as needed
ax.set_axis_off()

# Create custom legend elements
legend_elements = [Line2D([0], [0], color=color, lw=2) for color in colors]

# Create the legend
legend = ax.legend(legend_elements, labels, 
                   loc='center', ncol=2,
                   handler_map={Line2D: DoubleArrowHandler()},
                   handlelength=3)

# Adjust the layout
fig.tight_layout()

# Save the figure without extra white space
plt.savefig('/home/engaclew/agent/images/legend_art_params.png', bbox_inches='tight', pad_inches=0.1, dpi=300)
plt.show()
```
